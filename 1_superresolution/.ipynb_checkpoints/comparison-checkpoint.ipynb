```python
def compute_tke_spectrum_timeseries_torch(u_all, lx, ly, smooth=False, device='cpu'):
    u_all = torch.tensor(u_all, dtype=DTYPE, device=device).unsqueeze(1)  # [time, 1, nx, ny]
    ntime, nf, nx, ny = u_all.shape

    ntot = nx * ny

    # FFT
    uh_all = torch.fft.fftn(u_all, dim=(-2, -1)) / ntot  # [1, ntime, nx, ny]

    # Compute energy: sum of squares of real and imaginary parts
    energy = 0.5 * (uh_all.real**2 + uh_all.imag**2).sum(dim=1)  # [ntime, nx, ny]
    # energy = torch.sqrt(energy)

    # Wavenumber grids
    kx = torch.fft.fftfreq(nx, d=lx / nx, device=device)
    ky = torch.fft.fftfreq(ny, d=ly / ny, device=device)
    # kz = torch.fft.fftfreq(nz, d=lz / nz, device=device)
    KX, KY = torch.meshgrid(kx, ky, indexing='ij')
    K_mag = torch.sqrt(KX**2 + KY**2 )  # [nx, ny, nz]

    # Flatten for binning
    k_bins = torch.round(K_mag.flatten() / K_mag.max() * (nx // 2)).to(torch.int64)  # [npts]
    max_bin = k_bins.max().item() + 1

    wave_numbers = torch.arange(max_bin, device=device) * (2 * np.pi / ((lx + ly)/2))

    # Flatten energy: [ntime, npts]
    energy_flat = energy.view(ntime, -1)  # [ntime, npts]

    # Allocate spectrum
    tke_spectrum = torch.zeros(ntime, max_bin, device=device)

    # Vectorized binning using broadcasting + scatter_add
    for k in range(max_bin):
        mask = (k_bins == k)
        if mask.any():
            tke_spectrum[:, k] = energy_flat[:, mask].sum(dim=1)

    knyquist = (2 * np.pi / ((lx + ly )/2)) * min(nx, ny) / 2

    return knyquist, wave_numbers.detach().cpu().numpy(), tke_spectrum.detach().cpu().numpy()  # [ntime, maxbin]
```

```python
def smoothen(x):
    """
    Downsamples and then upsamples a 3D array [nt, nx, ny] using bicubic interpolation.
    
    Parameters:
    - x: np.ndarray of shape [nt, nx, ny]
    - k: int, downsampling factor

    Returns:
    - x_smooth: np.ndarray of shape [nt, nx, ny], smoothened version of x
    """
    bs, nt, nx, ny = x.shape

    # Downsample by factor k
    x_low = x[:, ::4, ::8, ::8]

    # Compute zoom factors: only zoom spatial dimensions
    zoom_factors = (1, nt / x_low.shape[1], nx / x_low.shape[2], ny / x_low.shape[3])

    # Upsample back using bicubic interpolation (order=3)
    x_smooth = zoom(x_low, zoom_factors, order=3)

    return x_smooth
```

```python
def compute_power(u_all, lx, ly, smooth=False, device='cpu'):
    bs, nt, nx, ny = u_all.shape
    u_all = u_all.reshape(-1, nx, ny)
    u_all = torch.tensor(u_all, dtype=DTYPE, device=device).unsqueeze(1)  # [time, 1, nx, ny]
    ntime, nf, nx, ny = u_all.shape

    ntot = nx * ny

    # FFT
    uh_all = torch.fft.fftn(u_all, dim=(-2, -1)) / ntot  # [1, ntime, nx, ny]

    # Compute energy: sum of squares of real and imaginary parts
    energy = 0.5 * (uh_all.real**2 + uh_all.imag**2).sum(dim=1)  # [ntime, nx, ny]
    # energy = torch.sqrt(energy)

    # Wavenumber grids
    kx = torch.fft.fftfreq(nx, d=lx / nx, device=device)
    ky = torch.fft.fftfreq(ny, d=ly / ny, device=device)
    # kz = torch.fft.fftfreq(nz, d=lz / nz, device=device)
    KX, KY = torch.meshgrid(kx, ky, indexing='ij')
    K_mag = torch.sqrt(KX**2 + KY**2 )  # [nx, ny, nz]

    # Flatten for binning
    k_bins = torch.round(K_mag.flatten() / K_mag.max() * (nx // 2)).to(torch.int64)  # [npts]
    max_bin = k_bins.max().item() + 1

    wave_numbers = torch.arange(max_bin, device=device) * (2 * np.pi / ((lx + ly)/2))

    # Flatten energy: [ntime, npts]
    energy_flat = energy.view(ntime, -1)  # [ntime, npts]

    # Allocate spectrum
    tke_spectrum = torch.zeros(ntime, max_bin, device=device)

    # Vectorized binning using broadcasting + scatter_add
    for k in range(max_bin):
        mask = (k_bins == k)
        if mask.any():
            tke_spectrum[:, k] = energy_flat[:, mask].sum(dim=1)


    knyquist = (2 * np.pi / ((lx + ly )/2)) * min(nx, ny) / 2

    return torch.log(tke_spectrum).detach().cpu().numpy()  # [ntime, maxbin]

def field_get_err(true, pred):
    return np.mean( np.mean( (true-pred)**2, axis=(2,3) )/np.mean(true**2, axis=(2,3)) )


def spec_get_err(true, pred):
    return np.mean( np.mean( (true-pred)**2, axis=(1,2)  ) /np.mean(true**2, axis=(1,2) ) )

def get_td(video):
    # video - [B,T,X,Y]
    td = video[:, 1:] - video[:,:-1]
    return td

def get_err(T, P, kind="minmax"):
    MIN = np.min(T) #min(np.min(T), np.min(P))
    MAX = np.max(T) #max(np.max(T), np.max(P))
    MEAN = np.mean(T)
    STD  = np.std(T)
    if kind=="minmax":
        shift = MIN
        scale = MAX-MIN
    elif kind=="meanstd":
        shift=MEAN
        scale=STD
    elif kind=="zeroone":
        shift=0
        scale=1
        
    T = (T-shift)/scale
    P = (P-shift)/scale
    
    # return np.mean(  np.sqrt(np.mean( (T-P)**2, axis=(2,3) ))  )
    return  np.mean( (T-P)**2 )**0.5

    

lx=3.2
ly=6.4

spec_true = compute_power(true[:,1:-1], lx, ly)
spec_pred_ls = []

for i in range(len(pred_ls)):
    spec_pred_ls.append(compute_power(pred_ls[i][:,1:-1], lx, ly))


field_err_ls = []
spec_err_ls = []


for i in range(len(pred_ls)):
    field_err_ls.append(get_err(true[:,1:-1], pred_ls[i][:,1:-1]))
    spec_err_ls.append(get_err(spec_true, spec_pred_ls[i]))
    
```

```python
# Your existing data setup
sample_id = 0
t_id = 2
true  = test_true[:, t_id] #[sample_id, :]  # assuming y_true is defined elsewhere
pred1 = no[:, t_id] #[sample_id, :]    # assuming y_no is defined elsewhere
pred2 = noadv[:, t_id] #[sample_id, :]  # assuming y_pred is defined elsewhere
pred3 = novae[:, t_id]
pred4 = nogan[:, t_id] #[sample_id, :]
pred5 = nodm[:, t_id]

t_ls = np.arange(true.shape[0])
time_ls = t_ls #np.linspace(0,10,11)[1:]

print(f"true: {true.shape}")
print(f"pred1: {pred1.shape}")
print(f"pred2: {pred2.shape}")

# Helper functions as you defined
def rel_l2(T, P):
    T = torch.tensor(T, dtype=torch.float32)
    P = torch.tensor(P, dtype=torch.float32)
    return (torch.norm(T-P, p=2) / torch.norm(T, p=2)).item()



# --- Figure & layout ---
fig = plt.figure(figsize=(16, 6), constrained_layout=True)
gs = fig.add_gridspec(
    nrows=2, ncols=4,
    width_ratios=[1, 1, 1, 0.05],  # last col for colorbar
    height_ratios=[1, 1]
)

axes = np.array([
    [fig.add_subplot(gs[0, 0]), fig.add_subplot(gs[0, 1]), fig.add_subplot(gs[0, 2])],
    [fig.add_subplot(gs[1, 0]), fig.add_subplot(gs[1, 1]), fig.add_subplot(gs[1, 2])]
])

cax = fig.add_subplot(gs[:, 3])  # colorbar axis spanning both rows

CMAP = 'gray'
vmin = min(true.min(), pred1.min(), pred2.min())
vmax = max(true.max(), pred1.max(), pred2.max())

titles = ["Ground Truth", "NO", "adv-NO", "NO + VAE", "NO + GAN", "NO + DM"]

images = []
for ax, title in zip(axes.ravel(), titles):
    im = ax.imshow(np.zeros_like(true[0]), origin='lower',
                   vmin=vmin, vmax=vmax, cmap=CMAP, aspect='auto')
    ax.set_title(title, fontsize=14, pad=4)
    ax.axis('off')
    images.append(im)

# One colorbar, created once:
cbar = fig.colorbar(images[0], cax=cax)
cbar.ax.tick_params(labelsize=10)

# Suptitle (constrained_layout handles room for it)
fig.suptitle("t: ", fontsize=24)


def update(frame_number):
    images[0].set_data(np.fliplr(true[frame_number]))
    images[1].set_data(np.fliplr(pred1[frame_number]))  # "NO"
    images[2].set_data(np.fliplr(pred2[frame_number]))  # "adv-NO"
    images[3].set_data(np.fliplr(pred3[frame_number]))  # "NO + VAE"
    images[4].set_data(np.fliplr(pred4[frame_number]))  # "NO + GAN"
    images[5].set_data(np.fliplr(pred5[frame_number]))  # "NO + DM"

    fig.suptitle(f"t: {time_ls[frame_number]}" + r"$\tau$", fontsize=24)
    return images


ani = FuncAnimation(fig, update, frames=len(t_ls), blit=True, repeat=True)
Writer = matplotlib.animation.writers['ffmpeg']
writer = Writer(fps=8, metadata=dict(artist='Me'), bitrate=1800)
ani.save('comparison_video_sch_imp_jet.mp4', writer=writer)
plt.close(fig)

```
